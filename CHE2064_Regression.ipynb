```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default= 'plotly_dark'
```

```python
x = np.linspace(-5,5,500)
y = np.random.normal(loc=2*x**3+3.5*x+1, scale=100)
```

```python
fig = make_subplots()
fig.add_scatter(x=x,y=y, mode='markers')
```

```python
A = np.c_[x**3, x**2, x, np.ones_like(x)]
A
```

```python
A.T @ A
v = np.linalg.solve(A.T @ A, A.T @ y)
y_model = A @ v
fig.add_scatter(x=x,y=y_model, mode='lines')
```

Fitting equation to points that lie on apparent circle

```python
N = 100
theta = np.random.uniform(0, 2*np.pi, N)
r = np.random.normal(3, 0.1, N)
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
fig2 = make_subplots()
fig2.add_scatter(x=x,y=y, mode='markers')
fig2.update_layout(width=600, height=600)
```

```python
A = np.c_[2*x, 2*y, np.ones_like(x)]
x0, y0, c = np.linalg.solve(A.T @ A, A.T @ (x**2 + y**2))
```

```python
x0, y0, c
rfit = np.sqrt(x0**2 + y0**2 + c)
rfit
```

```python
fig5 = make_subplots()
fig5.add_scatter(x=x,y=y, mode='markers')
fig5.update_layout(width=600, height=600)
```
