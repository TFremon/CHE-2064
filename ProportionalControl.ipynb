```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.integrate import solve_ivp
```

```python

def rhs(t, vec):
  h1, h2, h3 = vec

  kc = 1
 


  q_in = kc * (h3_set - h3)

  q_12 = np.sign(h1-h2) * a1 * np.abs(h1-h2)**(1/2)
  q_23 = np.sign(h2-h3) * a2 * np.abs(h2-h3)**(1/2)
  q_3 = a3*(h3)**(1/2)

  return (-q_12/A1, 
          (q_in + q_12 - q_23)/A2,
          (q_23 - q_3)/A3)
```

```python
t_in = 0
t_f = 200
h_i1, h_i2, h_i3 = 0.5, 1, 2
h_initial = [h_i1, h_i2, h_i3]
a1, a2, a3 = 0.1, 0.1, 0.1
A1, A2, A3 = 0.5, 1, 2
h3_set = 4
```

```python
sol = solve_ivp(rhs, (t_in, t_f), (h_initial), method = 'Radau', dense_output=True).sol
```

```python
t_plot = np.linspace(t_in, t_f, 1000)
h1, h2, h3 = sol(t_plot)
h3_setL = np.ones_like(t_plot)
h3_setL = [4 for i in h3_setL]

fig = make_subplots()
fig.add_scatter(x = t_plot, y= h1, name = 'h1')
fig.add_scatter(x = t_plot, y= h2, name = 'h2')
fig.add_scatter(x = t_plot, y= h3,name = 'h3')
fig.add_scatter(x = t_plot, y= h3_setL, name = 'Setpoint')
```
