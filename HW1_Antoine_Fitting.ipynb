```python
import numpy as np
from scipy.optimize import minimize
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default=('plotly_dark')
```

```python
data = np.loadtxt('WaterVaporPressure.csv', delimiter=',')
data
```

```python
T = data[:, 0]
log10P = np.log10(data[:, 1])
T, log10P
```

```python
def sse(v):
  a,b,c=v

  return np.sum((a-b/(c+T)-log10P)**2)
```

```python
g = [1., 2., 3.]
a,b,c=minimize(sse, g).x
a,b,c
```

```python
fig=make_subplots()
fig.add_scatter(x=T, y=log10P, mode='markers', name='Raw Data')
Tplot = np.linspace(273.15, 650, 1000)
fig.add_scatter(x=Tplot, y=a-b/(c+Tplot), mode='lines', name='Nonlinear regression' )
```

```python
Q = np.c_[T, -log10P, np.ones_like(T)]
h = T*log10P
a2, c2, d2 =np.linalg.solve(Q.T @ Q, Q.T @ h)
b2 = a2*c2 - d2

```

```python
fig2=make_subplots()
fig2.add_scatter(x=T, y=log10P, mode='markers', name='Raw Data')
Tplot = np.linspace(273.15, 650, 1000)
fig2.add_scatter(x=Tplot, y=a2-b2/(c2+Tplot), mode='lines', name='Linear regression' )
fig2.add_scatter(x=Tplot, y=a-b/(c+Tplot), mode='lines', name='Nonlinear regression' )
```
